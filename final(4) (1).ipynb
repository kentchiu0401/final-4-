```python
#在"大專校院校務資訊公開平臺"網站中要求使用者輸入學校，查詢該校的學生類人數資訊

import tkinter as tk
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


def search_school():
    school_name = entry_school_name.get()
    
    # 使用Selenium控制瀏覽器
    path = r'C:\Users\Evan\Downloads\anaconda\chromedriver-win64\chromedriver.exe'
    url='https://udb.moe.edu.tw/udata/'
    service=Service(executable_path=path)
    service
    # 使用 Chrome 開啟瀏覽器
    driver=webdriver.Chrome(service=service)
    driver.get(url)
    
    # 找到輸入框，輸入學校名稱，並按下查詢
    input_element = driver.find_element(By.ID, "KeyWord")
    input_element.send_keys(school_name)
    
    submit_button = driver.find_element(By.CLASS_NAME, "btn-default")
    submit_button.click()

    # 等待頁面載入
    WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, '//a[@title="' + school_name + '"]'))
    )

    # 找到學校名稱的連結，並點擊進入新頁面
    school_link = driver.find_element(By.XPATH, '//a[@title="' + school_name + '"]')
    school_link.click()
    
     # 等待頁面載入
    time.sleep(5)    
    #抓取學生類資料，其實就是函式search_student的內容
    try:
        # 使用Selenium的定位方法找到所有符合條件的元素
        elements = driver.find_elements(By.CLASS_NAME, 'fs-3')

        # 使用 for 迴圈處理每個元素
        for i in range(0, len(elements)):  # 從索引0開始
            try:
                # 取得文字內容
                text = elements[i].text
                # 輸出文字內容
                print(text)
            except Exception as e:
                print(f"Error while getting text: {e}")

    except NoSuchElementException as e:
        print(f"Element not found: {e}")

    finally:
        # 關閉瀏覽器
        driver.quit()

# 建立Tkinter視窗
root = tk.Tk()
root.title("學校查詢")

# 在視窗中加入Label和Entry供使用者輸入學校名稱
label_school_name = tk.Label(root, text="請輸入學校名稱:")
label_school_name.pack()
entry_school_name = tk.Entry(root)
entry_school_name.pack()

# 加入查詢按鈕，點擊後觸發search_school函數
search_button = tk.Button(root, text="查詢", command=search_school)
search_button.pack()

# 啟動Tkinter主迴圈
root.mainloop()
```
